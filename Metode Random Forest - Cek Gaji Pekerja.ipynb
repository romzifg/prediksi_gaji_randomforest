```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('gaji_train.csv')
df_test = pd.read_csv('gaji_test.csv')
```

```python
df_train.head(10)
```

```python
df_test.head(10)
```

# Preprocessing

```python
df_train = df_train.replace('?', np.nan)
df_test = df_test.replace('?', np.nan)
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train['Kelas Pekerja'] = df_train['Kelas Pekerja'].fillna(method='ffill')
df_test['Kelas Pekerja'] = df_test['Kelas Pekerja'].fillna(method='ffill')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df_train, y='Kelas Pekerja', hue='Gaji')
```

```python
df_train['Pekerjaan'] = df_train['Pekerjaan'].fillna(method='ffill')
df_test['Pekerjaan'] = df_test['Pekerjaan'].fillna(method='ffill')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df_train, y='Pekerjaan', hue='Gaji')
```

```python
df = [df_train, df_test]
for data in df:
    data['Keuntungan Kapital'] = data['Keuntungan Kapital'].astype('int')
    data['Kerugian Capital'] = data['Kerugian Capital'].astype('int')
    data['Jam per Minggu'] = data['Jam per Minggu'].astype('int')
    
for data in df:
    data.info()
```

```python
df_train['Pendidikan'].unique()
```

```python
jenjang = {'SD':1, '1st-4th':2, '5th-6th':3, '7th-8th':4,
              '9th':5, '10th':6, '11th':7, '12th':8, 'SMA':9,
              'Pendidikan Tinggi':10, 'D4':11, 'D3':12, 'Sarjana':13,
              'Master':14, 'Sekolah Professional':15, 'Doktor':16}

df_train['Pendidikan'] = df_train['Pendidikan'].replace(jenjang)
df_test['Pendidikan'] = df_test['Pendidikan'].replace(jenjang)
```

```python
df_train['Kelas Pekerja'].unique()
```

```python
classp = {'Wiraswasta':0, 'Pemerintah Lokal':1, 'Pekerja Bebas Perusahaan':2,
       'Pemerintah Negara':3, 'Pekerja Bebas Bukan Perusahan':4,
       'Pemerintah Provinsi':4, 'Tidak Pernah Bekerja':5, 'Tanpa di Bayar':6}

df_train['Kelas Pekerja'] = df_train['Kelas Pekerja'].replace(classp)
df_test['Kelas Pekerja'] = df_test['Kelas Pekerja'].replace(classp)
```

```python
gaji = {'<=7jt': 0, '>7jt':1}

df_train['Gaji'] = df_train['Gaji'].replace(gaji)
```

```python
cols = ['Jmlh Tahun Pendidikan', 'Status Perkawinan', 'Jenis Kelamin', 'Keuntungan Kapital', 'Kerugian Capital']

df_train.drop(cols, axis=1, inplace=True)
```

```python
df_train.head(10)
```

```python
# data train
x = df_train.drop(['Gaji'], axis=1)
y = df_train['Gaji']

x.head()
```

```python
x = pd.get_dummies(x, columns=['Pekerjaan'])
x.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
modell = RandomForestClassifier()
```

```python
from sklearn.metrics import accuracy_score
```

```python
model_rfc = RandomForestClassifier(criterion = 'entropy', n_estimators= 14, max_features = 'auto')
model_rfc.fit(X_train, y_train)
```

```python
y_predict_train = model_rfc.predict(X_test)
```

```python
accuracy_score(y_test, y_predict_train)
```
